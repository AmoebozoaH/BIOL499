```python
import pandas as pd
import glob
import os
import re
```

```python
# import file
df = pd.read_excel("../data_file/TAP metadata366 Oct2025 for canada_withedit.xlsx")
# use to map accession to file name
map_table = pd.read_excel("../data_file/Look_up_table.xlsx")
# loading fast names
fasta_files = glob.glob("../Assembly/kraken_purifyed_illumina/*_SPN.fasta")  
```

```python
# define target columns
# included PCN, Ceftriaxone*, Erythromycin*, Clindamycin*, Vancomycin, Tetracycline*, Levofloxacin
# unit: (mcg/mL)
cols = [
    "accessionNum", "spn_sero",
    
    "pcnmic_pre_np", "pcnmic_np",
    
    "cefmic_pre_np", "cefmic_np",
    
    "erythmic_pre_np", "erythmic_np",
    "clindamic_pre_np", "clindamic_np",
    "amoxmic_pre_np", "amoxmic_np",
    "methomic_pre_np", "methomic_np",
    "levomic_pre_np", "levomic_np",
    "pcn_sens_np",
    "pcnmic_pre_ear", "pcnmic_ear",
    "cefmic_pre_ear", "cefmic_ear",
    "erythmic_pre_ear", "erythmic_ear",
    "clindamic_pre_ear", "clindamic_ear",
    "amoxmic_pre_ear", "amoxmic_ear",
    "methomic_pre_ear", "methomic_ear",
    "levomic_pre_ear", "levomic_ear",
    "pcn_sens_ear", "ageInMonths",
]

# obtain a table with sub-column
df_sub = df[cols]
```

```python
# select target samples
target_accessions = ["W17143487", "F16380623", "W17653277", "M17701635", "T18271326", "M18434558", "F17654457", "F17896493",
    "F18239423", "T19376701", "M19156630", "T19510623", "M19257038", "F18448840", "F18448806", "H19271491",
    "T19641883", "H19332824", "T19860215", "W19724538", "W19756047", "W20312234", "M20130536", "H20143631",
    "H20175295", "T20595649", "H18637568", "H19691888", "H18510890", "M17945178", "F18442395", "M19219775",
    "T17202514", "M17112142", "H17221555", "M18000909", "W18629076", "W18265095", "H18485654", "M18587789",
    "M18617870",
]

# obtain target rows
df_filtered = df_sub[df_sub["accessionNum"].isin(target_accessions)]
```

```python
df_filtered
```

```python
# use to build join table 
# Build a dict[ our_number : filename ]
num_to_id = {}
for f in fasta_files:
    # obtain file
    fname = os.path.basename(f)
    # if file is target (according to regex)
    m = re.match(r"^(\d+)_.*_SPN\.fasta$", fname)
    if m:
        # obtain internal id
        our_num = int(m.group(1))
        fasta_id = os.path.splitext(fname)[0]
        num_to_id[our_num] = fasta_id

map_table["fasta_id"] = map_table["Our number"].map(num_to_id)
#map_table.to_excel("../data_file/map_id_fasta.xlsx", index=False)
```

```python
map_table
```

```python
map_table_filtered = map_table.drop(columns=["Our number", "Column1", "have data", "found on table"])
map_table_filtered = map_table_filtered.rename(columns={"Accession No": "accessionNum"})
```

```python
map_table_filtered
```

```python
merge_clinical = pd.merge(map_table_filtered, df_filtered, on="accessionNum", how="inner")
```
